# 001 - Basic LLM Chat

This notebook demonstrates the most basic interaction with a Large Language Model using Google's Gemini API.

## Setup

First, we need to import the necessary libraries and configure the API.

```python
import os

import google.generativeai as genai
from dotenv import load_dotenv
```

## Load Environment Variables

Load the API key from environment variables. Make sure you have a `.env` file with your `GOOGLE_API_KEY`.

```python
# Load environment variables from .env file
load_dotenv()

# Get API key from environment variable
api_key = os.getenv('GOOGLE_API_KEY')
if not api_key:
    raise ValueError("GOOGLE_API_KEY not found in environment variables. Please check your .env file.")
```

## Configure the Model

Configure the Gemini API and create a model instance.

```python
genai.configure(api_key=api_key)

model = genai.GenerativeModel("models/gemini-2.0-flash-lite")
```

## Simple Chat Example

Let's ask the model a simple question and see the response.

```python
response = model.generate_content("What is the capital of France?")
print(response.text)
```
