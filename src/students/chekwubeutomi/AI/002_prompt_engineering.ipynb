# 002 - Prompt Engineering

This notebook demonstrates how prompt quality affects AI responses. We'll compare vague prompts with specific, well-crafted prompts.

## Setup

Import libraries and configure the API.

```python
import google.generativeai as genai
import os
from dotenv import load_dotenv

# Load environment variables from .env file
load_dotenv()

# Get API key from environment variable
api_key = os.getenv('GOOGLE_API_KEY')
if not api_key:
    raise ValueError("GOOGLE_API_KEY not found in environment variables. Please check your .env file.")

# Configure the API
genai.configure(api_key=api_key)
model = genai.GenerativeModel("models/gemini-2.0-flash-lite")
```

## Helper Function

Create a helper function to display prompts and responses clearly.

```python
def ask_model(prompt):
    """Helper function to ask the model and display results"""
    print(f"Prompt: {prompt}")
    print("-" * 60)
    
    response = model.generate_content(prompt)
    print(f"Response:\n{response.text}")
    
    print("-" * 60)
```

## Example 1: Vague vs Specific

Let's compare a vague prompt with a specific one to see the difference in output quality.

```python
# Vague prompt
ask_model("Write about dogs")
```

```python
# Specific prompt with clear requirements
ask_model("Write a 100-word paragraph about the top 3 health benefits of owning a dog, with specific examples.")
```

## Example 2: Context Matters

Let's see how providing context about the audience improves the response.

```python
# No audience specified
ask_model("Explain machine learning")
```

```python
# Clear audience and constraints
ask_model("Explain machine learning to a 12-year-old using simple analogies. Keep it under 150 words.")
```

## Key Takeaways

- **Be specific**: Vague prompts lead to generic responses
- **Define your audience**: Tailor the complexity and style
- **Set constraints**: Word limits, format requirements, etc.
- **Provide context**: Help the model understand what you need
